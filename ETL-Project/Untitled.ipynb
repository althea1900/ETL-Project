```python
import pandas as pd
from sqlalchemy import create_engine
from Config import password
```

```python
worship_csv = "Resources/All_Places_of_Worship.csv"
worship_df = pd.read_csv(worship_csv)

gun_csv = "Resources/gun-violence-data_01-2013_03-2018.csv"
gun_df = pd.read_csv(gun_csv)

states_csv = "Resources/states.csv"
states_df = pd.read_csv(states_csv)
```

```python
states_df
```

```python

gun_tab = gun_df[["state", "n_killed", "n_injured"]].copy()
gun_tab = gun_tab.groupby("state").sum()
gun_tab
```

```python
worship_tab = worship_df[["Abbreviation", "ID"]].copy()
worship_tab = worship_tab.groupby("Abbreviation").count()
```

```python
full_df = gun_tab.merge(states_df, on = 'state')
full_df.set_index("Abbreviation")

full_df = full_df.merge(worship_tab, on = "Abbreviation")

full_df = full_df.rename(columns = {"ID":"Worship_Numbers"})
full_df.set_index("Abbreviation")

full_df["Total_Average"] = (full_df.n_killed+full_df.n_injured)/full_df["Worship_Numbers"]
full_df["Average_Injured"] = (full_df.n_injured)/full_df["Worship_Numbers"]
full_df["Average_Killed"] = (full_df.n_killed)/full_df["Worship_Numbers"]
full_df
```

```python
full_df.to_csv('Final.csv')
```
